### What is the top need for the top 10 race/ethnicities For each of them, what is the top need category?

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.options.display.max_columns = 50
# You'll probably need to run 'sudo pip install sdipylib' for this to work
from  sdipylib.url import cache_url
from IPython.core.display import HTML
print "pandas",pd.__version__
print "numpy", np.__version__
```

```python
fn  = cache_url('http://warehouse.sandiegodata.org/211/fc22cee6-acfa-4236-a6cb-e0d887eba178/211_deduped_service_calls.csv')
fn2 = cache_url('http://warehouse.sandiegodata.org/211/fc22cee6-acfa-4236-a6cb-e0d887eba178/211_need_tax.csv')
deduped = pd.read_csv(fn,dtype =   {15: str, 25: str, 27: str})
needs = pd.read_csv(fn2)

#Setting Index for DEDUPED
deduped = deduped.set_index(['id'])

#Setting Index for NEEDS
needs = needs.drop('id', 1)
needs = needs.set_index(['calls_id'])
needs.index.names = ['id']

deduped = needs.join(deduped, how='left')
```

```python
race_remove = ['Call disconnected', 'Not appropriate for this call', 'Other', 'Declined to State' 
               'Did not ask', 'Call dropped','Biracial/multi-ethnic', 'nan']
race = deduped
race = race.drop('ethnicity', 1)
race = race[race.race.notnull()]
for remove in race_remove:
    race = race[race.race != remove]

def extract_race_first_string(race_string):
    race_string_split = str(race_string).split(' - ')    
    return race_string_split[0]

def extract_race_second_string(race_string):
    race_string_split = str(race_string).split(' - ')    
    if len(race_string_split) == 2:
        return race_string_split[1]
    else:
        return None

race['ethnicity'] = race.race.apply(extract_race_first_string)
race['original_country'] = race.race.apply(extract_race_second_string)
```

```python
grouped_ethnicity = race.groupby('ethnicity')
need_count_ethnicity = grouped_ethnicity.need_tax_cat.value_counts().order(ascending=False)[:10]
need_count_ethnicity.plot(kind="barh", title='Top 10 needs by ethnicity')
```

```python
grouped_country = race.groupby('original_country')
need_count_country = grouped_country.need_tax_cat.value_counts().order(ascending=False)[:10]
need_count_country.plot(kind="barh", title='Top 10 needs by country of origin')
```
